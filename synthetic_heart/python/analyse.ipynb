```python
score_cols = []
if "Age" in df.columns: score_cols.append("Age")
if "Cholesterol_Total" in df.columns: score_cols.append("Cholesterol_Total")
if "Systolic_BP" in df.columns: score_cols.append("Systolic_BP")
if len(score_cols):
    df["risk_score"] = df[score_cols].apply(lambda row: row.sum()/len(score_cols), axis=1)
    display(df[["risk_score"]].describe())
```

```python
clean_path = "../data/heart_cleaned.csv"
df.to_csv(clean_path, index=False)
print("Saved cleaned CSV to", clean_path)

conn = sqlite3.connect("../data/heart.db")
df.to_sql("heart_data", conn, if_exists="replace", index=False)
print("Saved table 'heart_data' into SQLite DB at ../data/heart.db")
conn.close()
```

SQL QUERIES

```python
query = """
SELECT Heart_Disease, COUNT(*) AS count
FROM heart_data
GROUP BY Heart_Disease;
"""

pd.read_sql(query, conn)
```

```python
pd.read_sql("""
SELECT Heart_Disease,
       COUNT(*) * 100.0 / (SELECT COUNT(*) FROM heart_data) AS pct
FROM heart_data
GROUP BY Heart_Disease;
""", conn)
```

```python
pd.read_sql("""
SELECT Age, AVG(Cholesterol_Total) AS avg_chol
FROM heart_data
GROUP BY Age
ORDER BY Age;
""", conn)
```

```python
pd.read_sql("""
SELECT Heart_Disease, AVG(Systolic_BP) AS avg_sysbp
FROM heart_data
GROUP BY Heart_Disease;
""", conn)
```

```python
pd.read_sql("""
SELECT *
FROM heart_data
WHERE Cholesterol_Total IS NOT NULL
ORDER BY Cholesterol_Total DESC
LIMIT 10;
""", conn)
```

```python
pd.read_sql("""
SELECT Gender, Heart_Disease, COUNT(*) AS cnt
FROM heart_data
GROUP BY Gender, Heart_Disease
ORDER BY Gender, Heart_Disease;
""", conn)
```

```python
conn.execute("DROP VIEW IF EXISTS high_risk;")

conn.execute("""
CREATE VIEW high_risk AS
SELECT *, (Age + IFNULL(Cholesterol_Total, 0) + IFNULL(Systolic_BP,0))/3.0 AS simple_risk
FROM heart_data
WHERE Age IS NOT NULL;
""")

conn.commit()
print("View 'high_risk' created.")
```

```python
pd.read_sql("""
SELECT * FROM high_risk
ORDER BY simple_risk DESC
LIMIT 20;
""", conn)
```
